```python
 feature_differences = {}
```

```python
for feature in X_train:
        # Exclude the current feature
        X_train_reduced = X_train.drop(columns=[feature])
        X_val_reduced = X_val.drop(columns=[feature])

        # Train a new model without the current feature
        model_reduced = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state=42)
        model_reduced.fit(X_train_reduced, y_train)
        y_pred_reduced = model_reduced.predict(X_val_reduced)

        # Calculate the accuracy without the current feature
        reduced_accuracy = accuracy_score(y_val, y_pred_reduced)

        # Calculate the difference in accuracy
        accuracy_difference = original_accuracy - reduced_accuracy
        feature_differences[feature] = accuracy_difference

        print(f"Accuracy without '{feature}': {reduced_accuracy:.4f} (difference: {accuracy_difference:.4f})")

    # Find the least useful feature (smallest difference in accuracy)
        least_useful_feature = min(feature_differences, key=feature_differences.get)

        print(f"\nLeast useful feature: {least_useful_feature}")
        
```
